```python
df_adventure = pd.read_excel(r'C:\Users\Erica Rafael\Desktop\Unimed Ciência de Dados\Python para DS\AdventureWorks.xlsx',decimal='.')
```

```python
df_adventure.head()
```

```python
df_adventure.shape
```

```python
df_adventure.dtypes
```

```python
# receita total

df_adventure["Valor Venda"].sum()
```

```python
df_adventure['Custo Total'] = df_adventure['Custo Unitário'].mul(df_adventure["Quantidade"])
```

```python
df_adventure.head()
```

```python
# custo total

df_adventure['Custo Total'].sum()
```

```python
# lucro total

df_adventure['Lucro Total'] = df_adventure['Valor Venda'] - df_adventure["Custo Total"]
round(df_adventure['Lucro Total'].sum(),2)
```

```python
df_adventure["Tempo Envio"] = df_adventure['Data Envio'] - df_adventure['Data Venda']
```

```python
df_adventure.head()
```

```python
# extraindo apenas a parte numérica da duração em dias

df_adventure['Tempo Envio'] = (df_adventure['Tempo Envio']).dt.days
```

```python
df_adventure['Tempo Envio'].dtype
```

```python
# média de tempo de envio por marca

df_adventure.groupby('Marca')['Tempo Envio'].mean()
```

```python
# valores ausentes

df_adventure.isnull().sum()
```

```python
# lucro por ano e marca - index vem como o ano

df_adventure.groupby([df_adventure["Data Venda"].dt.year, "Marca"])['Lucro Total'].sum().unstack()
```

```python
# resetando index

df_lucro_ano = df_adventure.groupby([df_adventure["Data Venda"].dt.year, "Marca"])['Lucro Total'].sum().reset_index()
df_lucro_ano
```

```python
# total de vendas por produtos, da maior pra menor quantidade

df_adventure['Quantidade'].groupby(df_adventure['Produto']).sum().sort_values(ascending=False)

# mesmo resultado: 

df_adventure.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
df_adventure.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).plot.barh(title="Quantidade Produtos Vendidos")
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df_adventure.groupby(df_adventure['Data Venda'].dt.year)['Lucro Total'].sum().plot.bar(title="Lucro X Ano")
plt.xticks(rotation = "horizontal") 
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
# Selecionando apenas as vendas de 2009:

df_2009 = df_adventure[df_adventure["Data Venda"].dt.year == 2009]         
df_2009.head(5)
```

```python
# Printando um gráfico do Lucro X Mês dessa vendas:

df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro Total"].sum().plot(title = "Lucro X Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro")
plt.show()
```

```python
# Printando um gráfico do Lucro X Marca dessa vendas:

df_2009.groupby("Marca")["Lucro Total"].sum().plot.bar(title = "Lucro X Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal")                     
```

```python
# Printando um gráfico do Lucro X Classe dessa vendas:
df_2009.groupby("Classe")["Lucro Total"].sum().plot.bar(title = "Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal")                  
plt.show()
```
